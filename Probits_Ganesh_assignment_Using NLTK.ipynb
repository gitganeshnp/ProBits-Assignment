```python
import docxpy
import re
from docx import *
import pandas as pd
import os
import numpy as np
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from itertools import chain
```

```python
#Collected the finite skill set infromation

sk_set=pd.read_csv('C:\\Users\\Administrator\\Probits\\Final\\Data\\skills_set.csv')
skills_set=sk_set.skill_set.apply(lambda x:str(x).lower())
skills_set=skills_set.tolist()
skills_set
```

```python
#Processing the text from document and making it plain(Without any special characters)

text = docxpy.process('C:\\Users\\Administrator\\Probits\\Resumes\\Ganesh_Resume.docx')
res1 = " ".join(re.findall("[a-zA-Z0-9]+", text)) #retain only characters and digits
res1
```

```python
#Reomving stop wards function

stop = set(stopwords.words('english'))

def sw(text):
    text=[word for word in text.split() if word not in  stop]
    return " ".join(text)
```

```python
#cleaning text by converting to lower and producing word tokens

clean_text=sw(res1)
clean_text=clean_text.lower()
sentences = nltk.sent_tokenize(clean_text) 
words = [nltk.word_tokenize(sent) for sent in sentences] 

wor=list(chain.from_iterable(words))
wor
```

```python
#Getting the bigrams

bigrm = list(nltk.bigrams(clean_text.split()))
bigrm
```

```python
#Getting the skills & technologies from unigram 

output_list=[]
for w in wor:
    #print(w)
    if w in skills_set:   #checking unigram word present in skills_set
        if w not in output_list:  #inserting unique skills in the output list
            output_list.append(w)

output_list       
```

```python
#Getting the skills & technologies from bigram

for bi in bigrm:
    if ' '.join(bi) in skills_set:          #converting tuple to a string and checking in skills_set
        if ' '.join(bi) not in output_list:  #inserting unique skills in the output list
            output_list.append(' '.join(bi))
            
output_list        
```

```python
#Print the skills & technologies from the resume
output_list
```

# Converting as Structured data

```python

output_list_df = pd.DataFrame(output_list,columns=['SKills & Technologies'])
output_list_df
```

```python
output_list_df.to_csv("Output.csv",index = False)
```
